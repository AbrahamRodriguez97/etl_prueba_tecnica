```python
#Configuracion inical
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, lit, when

spark = SparkSession.builder.appName("ETL_Ventas").getOrCreate()
```

```python
#Extraccion 
ruta_entrada = "../data/ventas.csv"

df_ventas = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load(ruta_entrada)

df_ventas.show()
```

```python
#Transformacion

#Nulos
df_limpio = df_ventas.fillna({
    "cantidad": 0,
    "cliente_id": "Desconocido"
})

df_limpio = df_limpio.filter(col("fecha").isNotNull())

#Calcular el total de la venta
df_transformado = df_limpio.withColumn("total_venta", col("cantidad") * col("precio_unitario"))

df_transformado.show()
```

```python
#Carga
ruta_salida = "../data/ventas_transformado.csv"
df_transformado.write.mode("overwrite").csv(ruta_salida)
```
